```python
def f(x,y):
    return 25*(y+x**2)*(y+x**2) + (1 + x)*(1 + x)

x = np.arange(-3.0,3.0,0.1)
y = np.arange(-3.0,3.0,0.1)
X,Y = np.meshgrid(x,y)
Z = f(X,Y)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1,
                      cmap=plt.cm.RdBu, linewidth=0, antialiased=False)


ax.zaxis.set_major_locator(LinearLocator(6))

ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))


plt.show()
```

```python
import numpy as np

# Function to minimize
def f(x, y):
    return 25*(y+x**2)**2 + (1 + x)**2

def df(x, y):
    return 2*x*(50*y + 50*x**2 + 2) + 2 


# Gradient descent function
def gradient_descent(x, y, learning_rate, max_iterations):
    # Initialize variable to store the previous value of x
    prev_x = x
    prev_y = y
    
    # Loop for the maximum number of iterations
    for i in range(max_iterations):
        # Compute the gradient at x
        gradient_x = x
        gradient_y = y
        
        # Update x using the gradient and the learning rate
        x -= learning_rate * gradient_x
        y -= learning_rate * gradient_y
        
        # Check if the absolute difference between the function values at the previous and current x is less than the stop condition
        if abs(f(prev_x, prev_y) - f(x, y)) < 1e-6:
            break
        
        # Check if the absolute difference between the arguments at the previous and current x is less than the stop condition
        if abs(prev_x - x) < 1e-6:
            break
        
        # Check if te absolute difference between the first order derivatives at the previous and current x is less than the stop condition
        if abs(df(prev_x, prev_y) - df(x, y)) < 1e-6:
            break

        
        if abs(prev_y - y) < 1e-6:
            break
        
        if abs(df(prev_x, prev_y) - df(x, y)) < 1e-6:
            break
        
        # Update the previous value of x and y
        prev_x = x
    
    return x    
    return y
   
x = 10
y = 10

# Learning rates
learning_rates = [1e-6, 0.1, 1]

# Maximum number of iterations
max_iterations = [100, 1000]

# Loop over the learning rates and maximum number of iterations
for lr in learning_rates:
    for mi in max_iterations:
        # Perform gradient descent
        x_min = gradient_descent(x, y, lr, mi)
        y_min = gradient_descent(x, y, lr, mi)
        
        # Print the results
        print("Learning rate: {}".format(lr))
        print("Maximum number of iterations: {}".format(mi))
        print("Minimum value of x: {}".format(x_min))
        print("Minimum value of y: {}".format(y_min))
        print()
```
